```python
# Calculate percentiles
budget_threshold = main_data['Actual_Price'].quantile(0.25)
premium_threshold = main_data['Actual_Price'].quantile(0.75)

# Categorize products
def categorize_price(price):
    if price <= budget_threshold:
        return 'Budget'
    elif price <= premium_threshold:
        return 'Mid-Tier'
    else:
        return 'Premium'

main_data['Price_Category'] = main_data['Actual_Price'].apply(categorize_price)
```

```python
# Function for values in millions
def format_millions(value):
    if value >= 1e6:  
        return f"{value / 1e6:.1f}M" 
    elif value >= 1e3:  
        return f"{value / 1e3:.1f}K"
    elif value == 0:  # Handle zeroes explicitly
        return "0"
    else:
        return f"{value:.0f}"  

# Pivot table for heatmap
heatmap_data = regional_sales.pivot(index='Region_Group', columns='Price_Category', values='Total_Units_Sold')

# Fill NaN values with zeroes
heatmap_data = heatmap_data.fillna(0)

plt.figure(figsize=(8, 6))

# sns.heatmap
sns.heatmap(
    heatmap_data,
    annot=True,
    fmt=".0f",  # Adjust format for integers (if zeroes are not strings)
    cmap="Blues",
    linewidths=0.5,
    annot_kws={"fontsize": 10},
    cbar_kws={"format": "%.0f"}  
)

# Update annotations to use the format_millions function
for text in plt.gca().texts:  
    value = float(text.get_text())  
    text.set_text(format_millions(value))  

plt.title('Regional Sales Trends Across Price Categories')
plt.xlabel('Price Category')
plt.ylabel('Regions')
plt.tight_layout()
plt.show()
```

```python
# function to calculate and plot restricted trendlines
def add_trendline(ax, x, y, color, label):
    # Convert pandas Series to numpy arrays
    x = np.array(x)
    y = np.array(y)
    
    # Remove NaNs and zeros (log scale can't handle them)
    valid = (x > 0) & (y > 0)  # Ensure x and y are positive
    x, y = x[valid], y[valid]
    
    # Log-transform for log-log trendline
    x_log, y_log = np.log(x), np.log(y)
    
    # Fit regression line in log space
    coeffs = np.polyfit(x_log, y_log, deg=1)  # Linear regression on log-transformed data
    x_range = np.linspace(x.min(), x.max(), 100)  # Generate x values strictly within observed range
    trendline_range = np.exp(coeffs[1]) * x_range ** coeffs[0]

    # Plot the restricted trendline
    ax.plot(x_range, trendline_range, color=color, linestyle='--', label=label)
    print(f"{label}: y = {coeffs[0]:.2f} * log(x) + {coeffs[1]:.2f}")

# include only positive values for Actual_Price and Units_sold
filtered_data = main_data[(main_data['Actual_Price'] > 0) & (main_data['Units_sold'] > 0)]

# no negative values exist
print(f"Min Actual_Price: {filtered_data['Actual_Price'].min()}, Min Units_sold: {filtered_data['Units_sold'].min()}")

# Create side-by-side subplots for Ecological and Ethical products
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Ecological Plot
sns.scatterplot(ax=axes[0], data=filtered_data[filtered_data['Ecological'] == False], 
                x='Actual_Price', y='Units_sold', color='black', label='Non-Ecological', alpha=0.6)
sns.scatterplot(ax=axes[0], data=filtered_data[filtered_data['Ecological'] == True], 
                x='Actual_Price', y='Units_sold', color='green', label='Ecological', alpha=0.8)
# Add trendlines for Ecological
add_trendline(axes[0], filtered_data[filtered_data['Ecological'] == False]['Actual_Price'], 
              filtered_data[filtered_data['Ecological'] == False]['Units_sold'], 
              color='black', label='Trend (Non-Ecological)')
add_trendline(axes[0], filtered_data[filtered_data['Ecological'] == True]['Actual_Price'], 
              filtered_data[filtered_data['Ecological'] == True]['Units_sold'], 
              color='green', label='Trend (Ecological)')
# Set x and y limits based on the filtered data
axes[0].set_xlim(filtered_data['Actual_Price'].min(), filtered_data['Actual_Price'].max())
axes[0].set_ylim(filtered_data['Units_sold'].min(), filtered_data['Units_sold'].max())
axes[0].set_xscale('log')
axes[0].set_yscale('log')
axes[0].set_title('Ecological vs Non-Ecological Products')
axes[0].set_xlabel('Actual Price (Log Scale)')
axes[0].set_ylabel('Units Sold (Log Scale)')
axes[0].legend()

# Ethical Plot
sns.scatterplot(ax=axes[1], data=filtered_data[filtered_data['Ethical'] == False], 
                x='Actual_Price', y='Units_sold', color='black', label='Non-Ethical', alpha=0.6)
sns.scatterplot(ax=axes[1], data=filtered_data[filtered_data['Ethical'] == True], 
                x='Actual_Price', y='Units_sold', color='green', label='Ethical', alpha=0.8)
# Add trendlines for Ethical
add_trendline(axes[1], filtered_data[filtered_data['Ethical'] == False]['Actual_Price'], 
              filtered_data[filtered_data['Ethical'] == False]['Units_sold'], 
              color='black', label='Trend (Non-Ethical)')
add_trendline(axes[1], filtered_data[filtered_data['Ethical'] == True]['Actual_Price'], 
              filtered_data[filtered_data['Ethical'] == True]['Units_sold'], 
              color='green', label='Trend (Ethical)')
# Set x and y limits based on the filtered data
axes[1].set_xlim(filtered_data['Actual_Price'].min(), filtered_data['Actual_Price'].max())
axes[1].set_ylim(filtered_data['Units_sold'].min(), filtered_data['Units_sold'].max())
axes[1].set_xscale('log')
axes[1].set_yscale('log')
axes[1].set_title('Ethical vs Non-Ethical Products')
axes[1].set_xlabel('Actual Price (Log Scale)')
axes[1].set_ylabel('Units Sold (Log Scale)')
axes[1].legend()

plt.tight_layout()
plt.show()
```
